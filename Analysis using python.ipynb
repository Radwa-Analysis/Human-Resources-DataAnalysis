```python
#Concat the first name and the last name to the Full name and insert the new column at index 1  
new_col = employee_df['FirstName'] +' '+employee_df['LastName']
employee_df.insert(1,'FullName',new_col)
employee_df = employee_df.drop('FirstName',axis=1)
employee_df = employee_df.drop('LastName',axis=1)
employee_df
```

```python
#add DistanceLevel
employee_df['DistanceFromHome (KM)'] = pd.to_numeric(
    employee_df['DistanceFromHome (KM)'],
    errors='coerce'  )
bins = [-np.inf, 5, 15, 30, np.inf]
labels = ['Very Near', 'Near', 'Far', 'Very Far']

employee_df['DistanceLevel'] = pd.cut(
    employee_df['DistanceFromHome (KM)'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True)
```

```python
#Replace every state abbreviation by the main state name
employee_df['State'] = employee_df['State'].replace('NY', 'NewYork')
employee_df['State'] =employee_df['State'].replace('CA', 'California')
employee_df['State'] =employee_df['State'].replace('IL', 'Illinois')
employee_df
```

```python
#Create Age Range 
bins = [18, 30, 40, 50, 55]
labels = ['18-30', '31-40', '41-50', 'Above 50']
employee_df['Age Range'] = pd.cut(
    employee_df['Age'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True)
```

```python
#changing the data type of hiredate column
employee_df.HireDate = pd.to_datetime(employee_df.HireDate)
employee_df.info()
```

```python
#In Sales department changing job role manager to sales manager 
employee_df.loc[
    (employee_df["Department"] == "Sales") & 
    (employee_df["JobRole"].str.contains("Manager")), 
    "JobRole"
] = "Sales Manager"
       
```

```python
#Changing The department from technology to sales
employee_df.loc[
    (employee_df["Department"] == "Technology") & 
    (employee_df["JobRole"] == "Sales Executive"), 
    "Department"
] = "Sales"
```

```python
def get_salary_level(salary):
    if salary <= 50000:
        return 'Very Low'
    elif salary <= 120000:
        return 'Low'
    elif salary <= 250000:
        return 'Medium'
    elif salary <= 400000:
        return 'High'
    else:
        return 'Very High'
merged_employee_df['SalaryLevel'] = merged_employee_df['AnnualSalary'].apply(get_salary_level)
```
